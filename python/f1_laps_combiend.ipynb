```python
import pandas as pd

# Function to convert time string -> seconds (float)
def time_to_seconds(t):
    if pd.isna(t):  # missing values stay NaN
        return None
    parts = str(t).split(':')
    try:
        if len(parts) == 2:  # mm:ss.sss
            minutes, seconds = parts
            return int(minutes) * 60 + float(seconds)
        elif len(parts) == 3:  # hh:mm:ss.sss (rare, endurance sessions)
            hours, minutes, seconds = parts
            return int(hours) * 3600 + int(minutes) * 60 + float(seconds)
        else:
            return float(t)  # fallback if it's already numeric-like
    except:
        return None  # if parsing fails, return None

# Convert the lap and sector times into numeric versions
time_cols = ['LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time']

for col in time_cols:
    merged_df[col + '_s'] = merged_df[col].apply(time_to_seconds)

# Quick check
print(merged_df[[*time_cols, *(c + "_s" for c in time_cols)]].head())
```

```python
import os
import pandas as pd
import fastf1 as f1

# 0) make a cache folder first (fixes NotADirectoryError)
CACHE_DIR = "./fastf1_cache"
os.makedirs(CACHE_DIR, exist_ok=True)
f1.Cache.enable_cache(CACHE_DIR)

# 1) what you already have in 2021
events_2021_in_df = set(merged_df.loc[merged_df["Season"] == 2021, "EventName"].unique())

# 2) canonical 2021 calendar from FastF1
schedule = f1.get_event_schedule(2021, include_testing=False)
expected_2021 = set(schedule["EventName"].tolist())

missing_2021 = sorted(expected_2021 - events_2021_in_df)
print("Missing 2021 events:", len(missing_2021), "\n", missing_2021)
```

```python
# Map event name -> round for 2021
name_to_round = dict(zip(schedule["EventName"], schedule["RoundNumber"]))

new_parts = []
for event_name in missing_2021:
    rnd = int(name_to_round[event_name])
    try:
        session = f1.get_session(2021, rnd, "R")
        session.load()
        laps = session.laps

        # Build a dataframe compatible with your columns
        part = pd.DataFrame({
            "Driver":         laps["Driver"].values,
            "DriverNumber":   laps["DriverNumber"].astype(str).values,
            "Team":           laps["Team"].values,
            "LapNumber":      laps["LapNumber"].values,
            "LapTime":        laps["LapTime"].astype(str).values,
            "Sector1Time":    laps["Sector1Time"].astype(str).values,
            "Sector2Time":    laps["Sector2Time"].astype(str).values,
            "Sector3Time":    laps["Sector3Time"].astype(str).values,
            "Stint":          laps["Stint"].values,
            "Compound":       laps["Compound"].values,
            "TyreLife":       laps["TyreLife"].values,
            "LapStartTime":   laps["LapStartTime"].astype(str).values,
            "Season":         2021,
            "Round":          rnd,
            "EventName":      event_name,
        })

        # Winner (abbreviation of P1 finisher)
        part["Winner"] = (
            session.results.iloc[0]["Abbreviation"] if not session.results.empty else pd.NA
        )

        new_parts.append(part)
        print(f"✓ Loaded {event_name}")

    except Exception as e:
        print(f"✗ Failed {event_name}: {e}")

# Concatenate new races (if any)
new_2021_df = pd.concat(new_parts, ignore_index=True) if new_parts else pd.DataFrame()

# Ensure all columns exist and dtypes are compatible before concat
for col in merged_df.columns:
    if col not in new_2021_df.columns:
        new_2021_df[col] = pd.NA
new_2021_df = new_2021_df[merged_df.columns]

merged_df = pd.concat([merged_df, new_2021_df], ignore_index=True)

# Re-check
print(sorted(merged_df.loc[merged_df["Season"] == 2021, "EventName"].unique()))
```

```python
import os, fastf1 as f1, pandas as pd
os.makedirs("./fastf1_cache", exist_ok=True)
f1.Cache.enable_cache("./fastf1_cache")

def season_gaps(df, year):
    have = set(df.loc[df["Season"]==year, "EventName"].dropna().str.strip().unique())
    sched = f1.get_event_schedule(year, include_testing=False)
    expect = set(sched["EventName"].tolist())
    return sorted(expect - have), sorted(have - expect)  # missing, extras

for yr in [2019, 2020, 2021]:
    miss, extra = season_gaps(merged_df, yr)
    print(f"\n{yr}: missing={len(miss)} -> {miss}")
    if extra:
        print(f"{yr}: unexpected names in DF ({len(extra)}): {extra}")
```
