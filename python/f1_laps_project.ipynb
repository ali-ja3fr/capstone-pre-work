```python
import pandas as pd
import numpy as np

time_cols = ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time", "LapStartTime"]

def format_td(td):
    if pd.isna(td):  # handle NaN
        return np.nan
    total_seconds = td.total_seconds()
    hours = int(total_seconds // 3600)
    minutes = int((total_seconds % 3600) // 60)
    seconds = int(total_seconds % 60)
    milliseconds = int(td.microseconds / 1000)
    
    if hours == 0:
        return f"{minutes}:{seconds:02}.{milliseconds:03}"
    else:
        return f"{hours}:{minutes:02}:{seconds:02}.{milliseconds:03}"

for col in time_cols:
    f1_df[col] = pd.to_timedelta(f1_df[col], errors="coerce")
    f1_df[col] = f1_df[col].apply(format_td)

print(f1_df[time_cols].head())

```
